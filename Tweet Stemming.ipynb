```python
data = pd.read_csv('C:/Users/deepa/Documents/Sem 6/Data Mining and Data Warehousing/Twitter Data New.csv')
```

```python
data
```

```python
from nltk import PorterStemmer

tokenized_tweet = data['Filtered Text'].apply(lambda x: x.split())
```

```python
ps = PorterStemmer()

tokenized_tweet = tokenized_tweet.apply(lambda x: [ps.stem(i) for i in x])

tokenized_tweet.head()
```

```python
for i in range(len(tokenized_tweet)):
    tokenized_tweet[i] = ' '.join(tokenized_tweet[i])

data['Clean Tweets'] = tokenized_tweet
data.head()
```

```python
df = pd.DataFrame(data)
```
